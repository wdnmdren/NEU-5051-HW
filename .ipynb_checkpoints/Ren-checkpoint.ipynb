**(a)** Since a valid swap in the algorithm must satisfy two conditions: 1.The new edges created after the swap cannot duplicate existing edges. 2.he new edges cannot introduce self-loops. This means an infinite loop can occur in cases like a complete graph: since every pair of nodes is already connected, no new edge combination can be added without duplicating existing edges.

**(b)**

```python
procedure EDGE EXCHANGE SAFE(G, t, max_attempts)
    s ← 0
    attempts ← 0  #NEW code here
    while s < t and attempts < max_attempts do
        (i, j) ← get random edge(G)
        (u, v) ← get random edge(G)
        if i=u or i=v or j=u or j=v then
            attempts ← attempts + 1 #NEW code here
            continue
        end if
        if has edge(G, i, v) or has edge(G, j, u) then
            attempts ← attempts + 1 #NEW code here
            continue
        end if
        remove edge(G, i, j)
        remove edge(G, u, v)
        add edge(G, i, v)
        add edge(G, j, u)
        s ← s + 1
        attempts ← 0 #NEW code here
    end while
    return G
end procedure
```

**(c)**
To maintain connectivity, we cannot perform a swap that splits the network into multiple connected components. we need to check whether the graph remains connected before and after the swap. After updating a connection, we need to check is_connected(G). If it's not connected, the operation must be undone.
Checking for connectivity or bridges requires additional computation, which increases complexity. The number of swappable edges decreases, making it harder to find a suitable swap and increasing the number of failed attempts.

```python
procedure EDGE EXCHANGE CONNECTED(G, t)
    s ← 0
    while s < t do
        (i, j) ← get random edge(G)
        (u, v) ← get random edge(G)
        if i = u or i = v or j = u or j = v then
            continue
        end if
        if has edge(G, i, v) or has edge(G, j, u) then
            continue
        end if
        remove edge(G, i, j)
        remove edge(G, u, v)
        add edge(G, i, v)
        add edge(G, j, u)
        if not is_connected(G) then #NEW code here
            remove edge(G, i, v)
            remove edge(G, j, u)
            add edge(G, i, j)
            add edge(G, u, v)
        continue
        end if
        s ← s + 1
        end while
        return G
end procedure
```

**(d)**
Since we need to preserve the attribute pairs on edges, we must ensure that any newly created edge also connects either nodes with the same attribute or nodes with different attributes, consistent with the original. This can be achieved by adding a condition that checks whether the attributes of the two endpoint nodes match the required pattern.

In the original edge exchange, we only needed to avoid self-loops and multiple edges. Now, with the additional requirement of matching attribute pairs, many candidate exchanges will be rejected, which makes the overall process take longer.

```python
procedure EDGE EXCHANGE WITH ATTRIBUTES(G, t, x)
s ← 0
while s < t do
    (i, j) ← get random edge(G)
    (u, v) ← get random edge(G)
    if i = u or i = v or j = u or j = v then
        continue
    end if
    if has edge(G, i, v) or has edge(G, j, u) then
        continue
    end if
    if (x[i], x[j]) ≠ (x[i], x[v]) or (x[u], x[v]) ≠ (x[j], x[u]) then #NEW code here
        continue   
    end if
    remove edge(G, i, j)
    remove edge(G, u, v)
    add edge(G, i, v)
    add edge(G, j, u)
    s ← s + 1
end while
return G
end procedure
```

## Question 3

**(a)** betweenness centrality

```python
import networkx as nx

def betweenness_centrality(G, node_i):
    nodes = list(G.nodes())
    centrality = 0.0

    nodes_copy = nodes[:]
    while nodes_copy:
        s = nodes_copy.pop(0)
        for t in nodes_copy:
            if node_i in (s, t):
                continue
            try:
                paths = list(nx.all_shortest_paths(G, source=s, target=t))
            except nx.NetworkXNoPath:
                continue
            sigma_st = len(paths)
            sigma_st_v = sum(1 for path in paths if node_i in path[1:-1])
            if sigma_st > 0:
                centrality += sigma_st_v / sigma_st
                
    return centrality
                
```

```python
import networkx as nx

def betweenness_centrality(G, node_i):
    if not isinstance(G, nx.Graph):
        raise TypeError("A graph that is not a networkx object") # ii
    if node_i not in G:
        raise ValueError("A node that is not in the graph") # i
    if len(G) == 0:
        raise ValueError("A graph with no connectivity") # iii
    if not nx.is_connected(G.to_undirected()):
        raise ValueError("A graph with no connectivity") # iii
        
    nodes = list(G.nodes())
    centrality = 0.0

    nodes_copy = nodes[:]
    while nodes_copy:
        s = nodes_copy.pop(0)
        for t in nodes_copy:
            if node_i in (s, t):
                continue
            try:
                paths = list(nx.all_shortest_paths(G, source=s, target=t))
            except nx.NetworkXNoPath:
                continue
            sigma_st = len(paths)
            sigma_st_v = sum(1 for path in paths if node_i in path[1:-1])
            if sigma_st > 0: # situation that might produce numerical difficulties # iiii
                centrality += sigma_st_v / sigma_st
                
    return centrality
                
```

## Question 4

The road-minnesota network is a road network of the state of Minnesota in the United States, where nodes represent intersections and edges represent road segments connecting them. It is an undirected and unweighted spatial network. https://networkrepository.com/road-minnesota.php

```python
import scipy.io
import networkx as nx
import matplotlib.pyplot as plt

file_path = 'road-minnesota.mtx'

sparse_matrix = scipy.io.mmread(file_path)

graph = nx.from_scipy_sparse_array(sparse_matrix)

plt.figure(figsize=(10, 10))

pos = nx.spring_layout(graph, seed=42, iterations=50) 

nx.draw(
    graph, 
    pos, 
    node_size=5,       
    width=0.5,          
    edge_color='gray',  
    with_labels=False  
)

plt.title(f"'{file_path}' ")
plt.show()

degree = [d for n, d in G.degree()]
betweenness = list(nx.betweenness_centrality(G).values())
clustering = list(nx.clustering(G).values())  # 或边权重等

fig, axs = plt.subplots(1, 3, figsize=(18, 5))
axs[0].hist(degree, bins=range(min(degree), max(degree)+2), color='blue', edgecolor='black')
axs[0].set_title('Degree Distribution')

axs[1].hist(betweenness, bins=20, color='red', edgecolor='black')
axs[1].set_title('Betweenness Centrality')

axs[2].hist(clustering, bins=20, color='green', edgecolor='black')
axs[2].set_title('Clustering Coefficient')

plt.suptitle("Distribution", fontsize=16)
plt.show()
```

```python
null_G = nx.configuration_model([d for n, d in G.degree()])
null_G = nx.Graph(null_G)
null_G.remove_edges_from(nx.selfloop_edges(null_G))

null_degree = [d for n, d in null_G.degree()]
null_betweenness = list(nx.betweenness_centrality(null_G).values())
null_clustering = list(nx.clustering(null_G).values())

fig, axs = plt.subplots(1, 3, figsize=(18, 5))
axs[0].hist(null_degree, bins=range(min(null_degree), max(null_degree)+2), color='skyblue', edgecolor='black')
axs[0].set_title('Degree Distribution (Null Model)')

axs[1].hist(null_betweenness, bins=20, color='salmon', edgecolor='black')
axs[1].set_title('Betweenness Centrality (Null Model)')

axs[2].hist(null_clustering, bins=20, color='lightgreen', edgecolor='black')
axs[2].set_title('Clustering Coefficient (Null Model)')

plt.suptitle("Null_graph Distribution", fontsize=16)
plt.show()
```

For the degree distribution, the original network and the null model are nearly identical, which is expected since the null model was constructed to preserve the degree distribution.

For betweenness centrality, the original network exhibits a heavily right-skewed distribution, with most nodes having values near zero and a few nodes showing significantly higher values. This suggests the presence of key bridge nodes that play an important role in facilitating communication across the network. In contrast, the null model’s distribution is more uniform, lacking a long tail, indicating a more random structure without prominent intermediary nodes.

For clustering coefficient, while most nodes in the original network still have low clustering, a small subset shows moderate values, indicating the presence of triangular structures or local communities. However, in the null model, clustering coefficients are almost entirely zero, as the random rewiring eliminates most closed triads.

These comparisons highlight that although the null model preserves the degree distribution, it fails to retain higher-order structural features like bridge nodes and local clustering, which are present in the original network.
